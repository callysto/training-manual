### 13. Best practices for code development

We will rely on Cybera personnel to guide us on this.

Some key ideas.

**Name spaces:** Python allows you to input many different packages of tools. It is fast but dangerous to import willy-nilly, like this
```
from numpy import *
```
This gives you access to functions like sin() and cos(), and many other functions, but the names may start to conflict if you load in too many packages like this.

Better to import using namespaces, like this
```
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
```
So now, you would call np.sin() to get the sine function from numpy. Note that sc.sin() is another version of the sine function, fortunately they both do the same thing. 

We will extend this list to a standard list of imports and their abbreviations.

**Grouping files:** If your demo has many file associated to it (images,source code, etc.) it is best to keep them all organized into folder and subfolders, so users know what "package" to copy. 

**Repect the repo:** We have a repo to place the demos. There is some organization there, so repect it. Put stuff where they belong.

You can create animated plots using the **animation** function from matplotlib. There are several steps. Once you have the appropriate libraries imported, you first set up a figure, define two helper funciton (init and animate) and then call the animation function which creates the sequence of frames to be displayed. 

An example is shown here. 

````python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
from matplotlib import animation, rc
from IPython.display import HTML
````

````python
# First set up the figure, the axis, and the plot element we want to animate
# Note this will display a blank graphic, would liek to get rid of this.
fig, ax = subplots()

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2);
````

````python
def init():
    line.set_data([], [])
    return (line,)

def animate(i):
    x = linspace(0, 2, 1000)
    y = sin(2 * pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

HTML(anim.to_html5_video())
````

Be patient. This may take a while to start.  

You may notice the image has a little controller at the bottom, to stop and start the animation.

### 14. Special topic: Animating graphs

We can use widgets to directly control graphics, rather than the animation technique described above.

Here is a simple example that plots $f(x) = x^k$ for various powers of k.

````python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
from ipywidgets import interact

x=linspace(0,1)

def update(k=0):
    plot(x,x**k)
    show()
    
interact(update,k=(1,10));   
````

Here's another example, where we adjust the amplitude, frequency, and phase of a sine wave.

````python
t = linspace(0,1,1000)

def plot_sine(Amplitude=4,freq=2,phase=0):
    plot(t,Amplitude*sin(2*pi*freq*t + phase))
    axis([0,1,-6,6])
    show()

interact(plot_sine,Amplitude=(0,10), freq=(1, 10), phase=(0,2*pi));
````

In the following code, we show how to plot a series of graphs in a simple loop. It is a little tricky. Some things to note:

- use the magic command "%matplotlib notebook" 
- the "%matplotlib inline" magic does not work here
- set up an initial figure using the subplots command
- make sure to set the axis limits that you want
- the main loop resets the y limit and draws
- include a sleep command, so the plotting pauses between each image

** WARNING*** Note that the notebook and inline versions are not compatible, so running one may mess up your other graphs. Also, this seems to fail in JupyterLab, due to some new design choices by the Jupyter people.

````python
%matplotlib notebook

from numpy import *
from matplotlib.pyplot import *
import time

x = linspace(0,1,1000)
y = 0*x

fig, ax = subplots()
line, = ax.plot(x,y)
ax.set_xlim(0,1)
ax.set_ylim(-1,1)


for k in range(1,10):
    line.set_ydata(sin(2*pi*k*x))
    fig.canvas.draw()
    fig.canvas.flush_events()
    time.sleep(1)
````

The following version seems to work in both Jupyter and Jupyter Lab.

````python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
import time

handle = display.display(None, display_id=True)

x = linspace(0,1,1000)
y = 0*x
fig, ax = subplots()

line, = ax.plot(x,y)
ax.set_xlim(0,1)
ax.set_ylim(-1,1)

for k in range(1,10): 
   line.set_ydata(sin(2*pi*k*x))
   handle.update(fig)    
   time.sleep(.1)
````

### 14. Special topics: Audio

The IPython package makes it easy to play out a sound. Some notes on this:
<https://ipython.org/ipython-doc/3/api/generated/IPython.display.html>

Here, I just create a sampled sine wave and pump it out the speaker.

````python
%matplotlib inline
from matplotlib.pyplot import *
from numpy import *
from IPython.display import Audio, display

max_time = 5
rate = 8000
t = linspace(0,max_time,rate*max_time)
signal = sin(2*pi*440*t)
display(Audio(data=signal, rate=rate,autoplay=True))
````

You can make it a little more interesting by adding an interactive widget, as in the following. (Note you have to click on the arrow to make the sound start.)

````python
%matplotlib inline
from matplotlib.pyplot import *
from numpy import *
from ipywidgets import interactive
from IPython.display import Audio, display

def beat_freq(f1=220.0, f2=224.0):
    max_time = 5
    rate = 8000
    times = linspace(0,max_time,rate*max_time)
    signal = sin(2*pi*f1*times) + sin(2*pi*f2*times)
    display(Audio(data=signal, rate=rate,autoplay=False))
    return signal

v = interactive(beat_freq, f1=(200.0,300.0), f2=(200.0,300.0))

display(v)
````

### 14. Special topic: Hacks to input html code and Javascript

The Jupyter server and iPython can easily display html code that is in a file. The commands for this are simply to import IFrame and then call it with the file name, and dimensions for the display:

```
from IPython.display import IFrame
IFrame('mb.html',500,500)
```

However, the Jupyter server is sometimes very cautious about Javascript code embedded in the html file, and will attempt to 'scrub' it, removing what it thinks is dangerous. A simple hack to avoid this is to re-name the file something other than .html, read it in, and write it again with the .hmtl suffix. Somehow this convinces Jupyter that the file is safe. Which is crazy, right?

Here is an example, reading in a file called "MovingBalls.template" and then saving it under a new name.

```
with open('MovingBalls.template','r') as f:
    s=f.read()
    with open('mb.html','w') as g:
        g.write(s)
    
IFrame('mb.html',500,500)

```

### 14. Special topic: Making a script in Mac OS X

I find it useful to have a small command file in my Code folder, that I double click and it launches Jupyter for me, localized to the folder I want. A useful solution is to create a simple text file that has a list of terminal commands and turn that into a clickable application. 

The steps are as follows:

- Create a text file with a name like my_script.command
- Start it with a "shebang" command: #!/bin/bash
- list the terminal commands you want to run
- save the file
- set executive permission (i.e. in the terminal, run chmod +x /path/to/my_script.command)

For instance, here is the text file I used to customize my Jupyter launch:

```
#!/bin/bash
cd Dropbox/PIMS_Innovation/2018_CanCode/CallystoCode
jupyter notebook
```

That's all you need. I am sure there are similar methods in UNIX or Windows.
